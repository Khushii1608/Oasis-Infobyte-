# TIME SERIES ANALYSIS

```python
# Load the breakfast nutrition data
breakfast_nutrition = pd.read_csv('breakfast_nutrition.csv', parse_dates=['Carbohydrates'], index_col='Carbohydrates')

# Print the first few rows of the data
breakfast_nutrition.head()
```

```python
#Average Monthly Nutrition
df['Sugars'] = pd.to_datetime(df['Sugars'])

# Resample the 'Total Fat' column monthly and calculate the mean
total_fat = df.resample('M', on='Sugars')['Total Fat'].mean()

# Plot the mean of monthly breakfast
total_fat.plot()
plt.xlabel('Sugars')
plt.ylabel('Mean Total Fat')
plt.title('Mean Monthly nutrition')
# plt.show()
```

```python
# This the how we create all boxplot with a single code
import seaborn as sns
def outliers(x):
    y=x.select_dtypes(include=[int,float])
    for i in y:
        sns.boxplot(x=df[i])
        plt.title(i)
        plt.show()    
```

```python
import pandas as pd 
df = pd.read_csv("breakfast_nutrition.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

COR_MAT = df[['Calories', 'Sugars', 'Protein']].corr()
# Set up the matplotlib figure
plt.figure(figsize=(6, 6))
# Create a heatmap using Seaborn
sns.heatmap(COR_MAT, annot=True, cmap='YlOrRd')
# Set the title of the plot
plt.title('Correlation of Customer Demographics')
# Show the plot
plt.show()
```
